```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = {
    'region': ['North', 'East', 'West', 'South', 'North', 'East', 'West', 'South'],
    'units_sold': [100, 150, 80, 120, 130, 170, 90, 140],
    'revenue': [1000, 1500, 800, 1200, 1300, 1700, 900, 1400],
    'cost': [700, 1100, 600, 1000, 800, 1300, 650, 1100],
    'category': ['Electronics', 'Clothing', 'Electronics', 'Clothing', 'Furniture', 'Furniture', 'Electronics', 'Clothing'],
    'date': pd.date_range(start='2023-01-01', periods=8, freq='M')
}
```

```python
df = pd.DataFrame(data)
```

```python
sns.set(style = 'whitegrid')
```

# Comparision Plots:

# Bar Plot

```python
plt.figure(figsize=(10,6))
sns.barplot(x='region', y='units_sold', data=df, palette='Blues_d')
plt.title('Units Sold per Region')
plt.xlabel('Region')
plt.ylabel('Units Sold')
plt.show()
```

# Line Plot

```python
plt.figure(figsize=(10,6))
sns.lineplot(x='date', y='revenue', data=df, marker='o', color='g')
plt.title('Revenue Trend Over Time')
plt.xlabel('Date')
plt.ylabel('Revenue')
plt.xticks(rotation=45)
plt.show()
```

# Scatter Plot

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x='revenue', y='cost', hue='region', size='units_sold', sizes=(50, 300), data=df, palette='deep')
plt.title('Revenue vs. Cost (By Region)')
plt.xlabel('Revenue')
plt.ylabel('Cost')
plt.show()
```

# Composition Plots:
 

# Pie Chart

```python
category_sales = df.groupby('category')['units_sold'].sum()
plt.figure(figsize=(8,8))
plt.pie(category_sales, labels=category_sales.index, autopct='%1.1f%%', colors=sns.color_palette('Set3'))
plt.title('Proportion of Sales by Category')
plt.show()
```

```python
df_grouped = df.groupby('region').agg({'revenue': 'sum', 'cost': 'sum'}).reset_index()
```

```python
fig, ax = plt.subplots(figsize=(10,6))
```

```python
ax.bar(df_grouped['region'], df_grouped['revenue'], label='Revenue', color='b')
```

```python
ax.bar(df_grouped['region'], df_grouped['cost'], bottom=df_grouped['revenue'], label='Cost', color='r')
```

```python
plt.title('Revenue and Cost Composition per Region')
plt.xlabel('Region')
plt.ylabel('Amount ($)')
plt.legend()
plt.show()
```
