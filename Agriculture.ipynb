```python
import matplotlib.pyplot as plt
```

```python
years = [2018, 2019, 2020, 2021, 2022]
rice_yield = [2.8, 3.1, 2.9, 3.4, 3.2]  
wheat_yield = [3.2, 3.5, 3.0, 3.8, 3.6]  
```

```python
plt.plot(years, rice_yield, label = 'Rice Yield', color = 'grey', marker = 'o')
```

```python
plt.plot(years, wheat_yield, label = 'Wheat Yield', color = 'gold', marker = 'o')
```

```python
plt.title('Crop Yield Over the Years (2018-2022)', fontsize=14)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Yield (Tons per Hectare)', fontsize=12)
```

```python
plt.legend()
```

```python
for i, txt in enumerate(rice_yield):
    plt.annotate(f'{txt}', (years[i], rice_yield[i]), textcoords="offset points", xytext=(0,5), ha='center')
for i, txt in enumerate(wheat_yield):
    plt.annotate(f'{txt}', (years[i], wheat_yield[i]), textcoords="offset points", xytext=(0,5), ha='center')    
```

```python
plt.grid(True)
```

```python
plt.show()
```
