```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('apple_stock_data.csv')
```

```python
print(data.head())
```

```python
data['Date'] = pd.to_datetime(data['Date'])
```

```python
closing_prices = data['Close']
```

```python
mean_price = np.mean(closing_prices)
median_price = np.median(closing_prices)
mode_price = closing_prices.mode()[0]  
std_dev_price = np.std(closing_prices)
```

```python
print(f"Mean Closing Price: {mean_price}")
print(f"Median Closing Price: {median_price}")
print(f"Mode Closing Price: {mode_price}")
print(f"Standard Deviation of Closing Prices: {std_dev_price}")
```

```python
plt.figure(figsize=(10, 5))
```

```python
plt.plot(data['Date'], closing_prices, label='Closing Price')
```

```python
plt.axhline(y=mean_price, color='r', linestyle='-', label=f'Mean: {mean_price:.2f}')
plt.axhline(y=median_price, color='g', linestyle='--', label=f'Median: {median_price:.2f}')
plt.axhline(y=mode_price, color='b', linestyle=':', label=f'Mode: {mode_price:.2f}')
```

```python
plt.title('Apple Stock Closing Prices')
plt.xlabel('Date')
plt.ylabel('Closing Price')
plt.legend()
```

```python
plt.xticks(rotation=45)
plt.tight_layout()
```

```python
plt.show()
```
