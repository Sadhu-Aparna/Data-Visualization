```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,10,100)
y = np.sin(x)
```

# Line Plot

```python
plt.figure(figsize=(10, 6))
plt.subplot(2, 3, 1)
plt.plot(x, y)
plt.title('Line Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
```

# Scatter Plot

```python
plt.subplot(2, 3, 2)
plt.scatter(x, y, color='blue')
plt.title('Scatter Plot')
```

# Bar Plot

```python
categories = ['A', 'B', 'C', 'D']
values = [4, 7, 1, 8]
plt.subplot(2, 3, 3)
plt.bar(categories, values, color='green')
plt.title('Bar Plot')
```

# Histogram

```python
data = np.random.randn(1000)
plt.subplot(2, 3, 4)
plt.hist(data, bins=30, color='purple', alpha=0.7)
plt.title('Histogram')
```

# Pie Chart

```python
sizes = [15, 30, 45, 10]
labels = ['A', 'B', 'C', 'D']
plt.subplot(2, 3, 5)
plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.title('Pie Chart')
```

```python
plt.tight_layout()
plt.show()
```
